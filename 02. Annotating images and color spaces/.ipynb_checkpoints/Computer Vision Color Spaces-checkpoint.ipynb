# Computer Vision color spaces

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
img=cv2.imread('purple-flowers.jpg')
rgb_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,20))
plt.imshow(rgb_img)
cop_img=img.copy()
```

[Hint](# "Numpy .copy function") 

## Color Spaces

### BGR Spliting the color channels

```python
# Load a color image and split in is different B, G and R channels
cop_img=img.copy()
b,g,r = cv2.split(cop_img)
```

[Hint](# "cv2.split") 

```python
# Show each channel channel on screen
b = cop_img[:,:,0]
g = cop_img[:,:,1]
r = cop_img[:,:,2]
cv2.imshow("Red", r)
cv2.imshow("Green", g)
cv2.imshow("Blue", b)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(r)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(g)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(b)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

#### Merging the color channels

```python
# Modify a channel and then merge them all again
zeros = np.zeros(cop_img.shape[:2], dtype="uint8")
cv2.imshow("Red", cv2.merge([zeros, zeros, r]))
cv2.imshow("Green", cv2.merge([zeros, g, zeros]))
cv2.imshow("Blue", cv2.merge([b, zeros, zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
