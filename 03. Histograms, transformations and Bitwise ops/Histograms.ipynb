# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
img=cv2.imread('day1.jpg')
w,h= img.shape[:2]
b_hist = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(b_hist,color='b')
```

```python
g_hist = cv2.calcHist([img],[1],None,[256],[0,256])
plt.plot(g_hist, color='g')
```

```python
r_hist = cv2.calcHist([img],[2],None,[256],[0,256])
plt.plot(r_hist,color='r')
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
hist1 = cv2.calcHist([img],[0],None,[256],[0,256])
hist2 = cv2.calcHist([img],[1],None,[256],[0,256])
hist3 = cv2.calcHist([img],[2],None,[256],[0,256])

plt.subplot(221), plt.imshow(img)
plt.subplot(222), plt.plot(hist1), plt.plot(hist2),plt.plot(hist3)
plt.xlim([0,256])

plt.show()
```
